Question 6: A shopkeeper wants to calculate total bill after discount.

Input: Enter price and discount percentage

Output: Final price after discount

```python
price = float(input("Enter the original price: "))
discount_percentage = float(input("Enter the discount percentage: "))

# Calculate final price directly
final_price = price * (1 - discount_percentage / 100)

print(f"The final price after a {discount_percentage}% discount is: ${final_price:.2f}")
```
