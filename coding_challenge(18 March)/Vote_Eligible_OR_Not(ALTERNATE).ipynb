Question 7: Check whether a person is eligible to vote.

Input: Enter age

Output: Eligible / Not Eligible

```python
def check_voting_eligibility(age):
    if age >= 18:
        return "Eligible"
    else:
        return "Not Eligible"

age = int(input("Enter age: "))
result = check_voting_eligibility(age)
print(result)
```
