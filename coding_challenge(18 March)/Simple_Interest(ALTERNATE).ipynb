Question 10: Create a function to calculate simple interest.

Input: Enter P, R, T

Output: Simple Interest

```python
def calculate_si(p, r, t):
    return (p * r * t) / 100

# Takes inputs separated by space
p, r, t = map(float, input("Enter P, R, T separated by space: ").split())
print(f"Simple Interest: {calculate_si(p, r, t)}")
```
